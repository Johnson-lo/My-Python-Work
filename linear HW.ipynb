```python
%matplotlib inline
import numpy as np
import sympy as sp 
import matplotlib.pyplot as plt
```

```python
t=np.linspace(-5,5,500)
x=t+1
y=3+t
```

```python
len(x)
```

```python
plt.plot(x,y)
```

```python
X=np.array([x,y])
```

```python
X.shape
```

```python
A=np.array([[1,2],[3,4]])
```

```python
egg=A@X
```

```python
egg.shape
```

```python
xi=egg[0]
yi=egg[1]
```

```python
plt.plot(xi,yi)
plt.xlim(-10,10)
plt.ylim(-10,10)
```

```python
plt.subplot(121)
plt.plot(x,y)
plt.xlim(-10,10)
plt.ylim(-10,10)
plt.subplot(122)
plt.plot(x,y)
plt.xlim(-10,10)
plt.ylim(-10,10)

```

```python
π=np.pi
θ=np.linspace(0,2*π)
x=np.cos(θ)
y=np.sin(θ)
plt.plot(x,y)
```

```python
ax=plt.gca()
plt.plot(x,y)
ax.set_aspect('equal')
```

```python
X=np.array([x,y])
egg=A@X
```

```python
xi=egg[0]
yi=egg[1]
```

```python
ax=plt.gca()
plt.plot(x,y)
ax.set_aspect
```

```python
plt.subplot(121)
ax=plt.gca()
plt.plot(x,y)
ax.set_aspect('equal')
plt.xlim(-5,5)
plt.ylim(-5,5)

plt.subplot(122)
ax=plt.gca()
plt.plot(xi,yi)
ax.set_aspect('equal')
plt.xlim(-5,5)
plt.ylim(-5,5)
```
