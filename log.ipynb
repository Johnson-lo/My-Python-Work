```python
%matplotlib inline 
import math
import matplotlib.pyplot as plt
import time
import numpy as np
```

```python
cost_time=[]
log_output=[]
for i in range(1,11):
    log_sum=0
    start=time.time()
    n = 10**7*i
    log_output.append(n)
    for j in range(1,n+1):
        a=math.log(j,10)
        log_sum+=a
    end=time.time()
    cost=end-start
    print(log_sum)
    print(cost)
    cost_time.append(cost)
plt.scatter(log_output,cost_time)
    
```
